```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('feature_enginering.csv')
```

```python
df.head()
```

```python
df['loan_status'].replace([1, 0], ['Approved', 'Rejected'], inplace=True)
```

```python
df.head()
```

```python
sns.countplot(x='person_gender', data=df, hue='loan_status', palette=['#e74c3c', '#2ecc71'])
plt.title('Loan Status According to Gender')
plt.savefig('loan_status_gender.png')
plt.show()
```

```python
sns.countplot(x='person_education', data=df, hue='loan_status', palette=['#e74c3c', '#2ecc71'])
plt.title('Loan Status According to Education Qualification')
plt.savefig('loan_status_education.png')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='loan_intent', data=df, hue='loan_status', palette=['#e74c3c', '#2ecc71'])
plt.title('Loan Status According to Loan Intent')
plt.savefig('loan_status_intent.png')
plt.show()
```

```python
sns.displot(df['person_age'], kde=True)
plt.title('Age')
plt.savefig('age_range.png')
plt.show()
```

```python
sns.displot(df['person_income'], kde=True)
plt.title('Income')
plt.savefig('income_range.png')
plt.ticklabel_format(style='plain', axis='x')
```

```python
sns.displot(df['loan_amnt'], kde=True)
plt.title('Loan Amount')
plt.savefig('loan_range.png')
plt.ticklabel_format(style='plain', axis='x')
```

```python
sns.displot(df['credit_score'], kde=True)
plt.title('Credit Score')
plt.savefig('credit_range.png')
plt.show()
```

```python
sns.scatterplot(x='person_income', y='loan_amnt', data=df, hue='loan_status', palette=['#2ecc71', '#e74c3c'])
plt.title('Income vs Loan Amount')
plt.savefig('income_vs_loan.png')
plt.ticklabel_format(style='plain', axis='x')
```
