```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('clean_data.csv')
```

```python
df.head()
```

```python
# set the value in percentage
df['loan_percent_income'] = df['loan_percent_income'] * 100
```

```python
df.head()
```

```python
# calclate the loan to income and income to ratio
df['loan_to_income'] = df['loan_amnt'] / df['person_income']
df['income_to_loan'] = df['person_income'] / df['loan_amnt']

# calcuate the left over income after loan cost
df['net_disposable_income'] = df['person_income'] - (df['loan_amnt'] * df['loan_int_rate'] / 100)
```

```python
df.to_csv('feature_enginering.csv', index=False)
```
